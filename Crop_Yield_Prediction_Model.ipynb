```python
df = df.drop(['Crop_Year','Fertilizer','Pesticide','Yield'], axis=1)
```

```python
def remove_outlier(col):

   sorted(col)

   q1,q3=col.quantile([0.25,0.75])

   IQR=q3-q1

   lwr_bound= q1-(1.5*IQR)

   upr_bound =q3+(1.5*IQR)

   return lwr_bound, upr_bound
```

```python
low, high=remove_outlier(df["Production"])
df["Production"]=np.where(df["Production"]>high, high, df["Production"])
df["Production"]=np.where (df["Production"]<low, low, df ["Production"])
df.boxplot(column="Production")
```

```python
low, high=remove_outlier(df["Area"])
df["Area"]=np.where(df["Area"]>high, high, df["Area"])
df["Area"]=np.where (df["Area"]<low, low, df ["Area"])
df.boxplot(column="Area")
```

```python
low, high=remove_outlier(df["Annual_Temp"])
df["Annual_Temp"]=np.where(df["Annual_Temp"]>high, high, df["Annual_Temp"])
df["Annual_Temp"]=np.where (df["Annual_Temp"]<low, low, df ["Annual_Temp"])
df.boxplot(column="Annual_Temp")
```

```python
low, high=remove_outlier(df["Annual_Rainfall"])
df["Annual_Rainfall"]=np.where(df["Annual_Rainfall"]>high, high, df["Annual_Rainfall"])
df["Annual_Rainfall"]=np.where (df["Annual_Rainfall"]<low, low, df ["Annual_Rainfall"])
df.boxplot(column="Annual_Rainfall")
```

## **One Hot Encoding and Standard Scaling**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

ohe = OneHotEncoder(drop='first', handle_unknown='ignore')  # Or 'use_encoded_value'
scale = StandardScaler()

preprocesser = ColumnTransformer(
    transformers=[
        ('StandardScale', scale, [3, 4, 5]),
        ('OHE', ohe, [0, 1, 2])
    ],
    remainder='passthrough'
)
```

```python
preprocesser
```

```python
X_train_dummy = preprocesser.fit_transform(X_train)
X_test_dummy = preprocesser.transform(X_test)
```

```python
X_train_dummy
```

```python
X_test_dummy
```

## **Modeling**

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

def calculate_adjusted_r2(y_true, y_pred, n_samples, n_features):

  r2 = r2_score(y_true, y_pred)
  return 1 - ((1 - r2) * (n_samples - 1) / (n_samples - n_features - 1))

models = {
  'lr': LinearRegression(),
  'lss': Lasso(),
  'Rid': Ridge(),
  'Dtr': DecisionTreeRegressor(),
  'gbr': GradientBoostingRegressor()
}

for name, md in models.items():
  md.fit(X_train_dummy, y_train)
  y_pred = md.predict(X_test_dummy)

  # Calculate adjusted R-squared
  adjusted_r2 = calculate_adjusted_r2(y_test, y_pred, X_test_dummy.shape[0], X_test_dummy.shape[1])

  print(f"{name} : mae : {mean_absolute_error(y_test, y_pred)} adjusted_r2 : {adjusted_r2}")
```

## **Select Model**

## **Prediction**

```python
def prediction(Crop, Season, State, Area, Annual_Rainfall, Annual_Temp):
    # Create an array of the input features
    features = np.array([[Crop, Season, State, Area, Annual_Rainfall, Annual_Temp]], dtype=object)

    # Transform the features using the preprocessor
    transformed_features = preprocesser.transform(features)

    # Make the prediction
    predicted_yield = dtr.predict(transformed_features).reshape(1, -1)

    return predicted_yield[0]

Crop= 'Arhar/Tur'
Season= 'Kharif'
State= 'Assam'
Area= 6637.0
Annual_Rainfall= 2051.4
Annual_Temp= 22.69


result = prediction(Crop,Season,State,Area,Annual_Rainfall,Annual_Temp)
```
